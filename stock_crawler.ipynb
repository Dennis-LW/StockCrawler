```python
import requests
from io import StringIO
import pandas as pd
import numpy as np
from tqdm import tqdm_notebook, tqdm
import time
import os



savePath = "./save/"

# 下載股價
def downloadStockToCsv(datestr,savePath):
    r = requests.post('https://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&date=' + datestr + '&type=ALL')
    if r.text:
        df = pd.read_csv(StringIO(r.text.replace("=", "")),
                         header=["證券代號" in l for l in r.text.split("\n")].index(True)-1)

        df.apply(lambda s: pd.to_numeric(s.astype(str).str.replace(",", "").replace("+", "1").replace("-", "-1"), errors='coerce'))
        df.to_csv(savePath+datestr+".csv", encoding='utf_8_sig')
    print("[*]"+datestr+"...OK")


#10 years data
for year in tqdm(range(2018,2019)):
    if year == 2021:
        for month in range(1,4):
            for day in range(1,32):
                if month < 10:
                    if day < 10:
                        datestr = str(year)+"0"+str(month)+"0"+str(day)
                        print("[*]Download..."+datestr)
                        downloadStockToCsv(datestr,savePath+str(year)+"/")
                        time.sleep(5)
                    else:
                        datestr = str(year)+"0"+str(month)+str(day)
                        print("[*]Download..."+datestr)
                        downloadStockToCsv(datestr,savePath+str(year)+"/")
                        time.sleep(5)
                else:
                    if day < 10:
                        datestr = str(year)+"0"+str(month)+"0"+str(day)
                        print("[*]Download..."+datestr)
                        downloadStockToCsv(datestr,savePath+str(year)+"/")
                        time.sleep(5)
                    else:
                        datestr = str(year)+"0"+str(month)+str(day)
                        print("[*]Download..."+datestr)
                        downloadStockToCsv(datestr,savePath+str(year)+"/")
                        time.sleep(5)
    else:
        for month in range(1,13):
            for day in range(1,32):
                if month < 10:
                    if day < 10:
                        datestr = str(year)+"0"+str(month)+"0"+str(day)
                        print("[*]Download..."+datestr)
                        downloadStockToCsv(datestr,savePath+str(year)+"/")
                        time.sleep(5)
                    else:
                        datestr = str(year)+"0"+str(month)+str(day)
                        print("[*]Download..."+datestr)
                        downloadStockToCsv(datestr,savePath+str(year)+"/")
                        time.sleep(5)
                else:
                    if day < 10:
                        datestr = str(year)+"0"+str(month)+"0"+str(day)
                        print("[*]Download..."+datestr)
                        downloadStockToCsv(datestr,savePath+str(year)+"/")
                        time.sleep(5)
                    else:
                        datestr = str(year)+"0"+str(month)+str(day)
                        print("[*]Download..."+datestr)
                        downloadStockToCsv(datestr,savePath+str(year)+"/")
                        time.sleep(5)

```
